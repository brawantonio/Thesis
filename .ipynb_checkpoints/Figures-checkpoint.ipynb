```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mi=np.random.rand(2, 500)
mf=0.5*(np.random.rand(2, 500)-0.5)
```

```python
plt.scatter(mi[0,:], mi[1,:], s=10, color='blue')
plt.quiver(mi[0,:], mi[1,:], mf[0,:], mf[1,:], alpha=1, color='lightseagreen')
plt.scatter(0.5, 0.5, s=7000, color='grey')

plt.axis('off')
plt.show()
plt.savefig('Brownian.png')
plt.close()
```
