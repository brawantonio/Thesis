```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
size = 100
t = np.linspace(0, 20, size)

xi = np.zeros(size)
x = np.zeros(size)
xi[int(0.1*size):int(0.15*size)]=1
xi[int(0.30*size):int(0.65*size)]=1
xi[int(0.85*size):size-1]=1
xi=2*xi-1

for i in range(size-1):
    x[i+1] = x[i]+xi[i]*0.1
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.step(t, xi)
plt.title(r'$\dot{x}(t) = \xi(t) = \pm c$', fontsize=18)
plt.ylabel(r'$\xi(t)/c $', fontsize=18)
plt.xlabel(r'$t$ [unidades arbitrarias]', fontsize=18)

plt.subplot(1, 2, 2)
plt.plot(t, x)
plt.title('$x(t)$', fontsize=18)
plt.xlabel(r'$t$ [unidades arbitrarias]', fontsize=18)

plt.savefig('dichotomous.png', bbox_inches='tight')
plt.show()
plt.close()
```

```python
points = np.array([0, 1, 2, 3, 4])
values = np.zeros(len(points))
```
