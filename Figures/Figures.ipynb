```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
bg = 100
mi=np.random.rand(2, bg)
mf=0.5*(np.random.rand(2, bg)-0.5)
```

```python
plt.scatter(mi[0,:], mi[1,:], s=10, color='black')
plt.quiver(mi[0,:], mi[1,:], mf[0,:], mf[1,:], alpha=1, color='grey',  width=0.010, scale=1/0.19)

plt.arrow(0.01, 0.7, 0.2, -0.5, head_width=0.025, head_length=0.05, fc='k', ec='k')
plt.arrow(0.21, 0.18, 0.6, 0.02, head_width=0.035, head_length=0.05, fc='k', ec='k')
plt.arrow(0.83, 0.22, -0.1, 0.6, head_width=0.035, head_length=0.05, fc='k', ec='k')
plt.arrow(0.72, 0.83, -0.10,-.13, head_width=0.033, head_length=0.05, fc='k', ec='k')

plt.scatter(0.5, 0.5, s=7000, color='orange')

plt.axis('off')
plt.savefig('Brownian.png', bbox_inches='tight')
plt.show()
plt.close()
```

```python
def f1(x):
    return 0.25*x**4 - 2*a*x**3/3.-0.5*x**2
def f2(x):
    return 0.25*x**4 + 2*a*x**3/3.-0.5*x**2
a= -0.25

A = a - np.sqrt(a**2+1)
C = a + np.sqrt(a**2+1)
B = 0
D = 0.5*(B+C)

points = np.array([A, B, C, D])
```

```python
x1 = np.linspace(-2.2, 1.9, 1000)
x2 = np.linspace(-2.2, 1.9, 1000)
x3 = np.linspace(B, D, 100)

plt.figure(figsize=(6,6))
plt.plot(x1, f1(x1), label=r'(a)', color='blue')
plt.plot(x2, f2(x2-2*a)+1, label=r'(b)', color='purple')
plt.plot(x3, f1(x3), color ='red')

plt.scatter(points, f1(points), color='blue')
plt.scatter(points, f2(points-2*a)+1, color='purple')

plt.text(points[0]+0.1, f1(points[0]+ 0.5), r'A', fontsize=18)
plt.text(points[1]+.1, f1(points[1]+ 1.2), r'B', fontsize=18)
plt.text(points[2]+.1, f1(points[2]+ 0.4), r'C', fontsize=18)
plt.text(points[3]+.1, f1(points[3]+ 0.8), r'D', fontsize=18)

plt.plot([points[0], points[0]], [f1(points[0]), f2(points[0]-2*a)+1], 'k--', lw=.8)
plt.plot([points[1], points[1]], [f1(points[1]), f2(points[1]-2*a)+1], 'k--', lw=.8)
plt.plot([points[2], points[2]], [f1(points[2]), f2(points[2]-2*a)+1], 'k--', lw=.8)
plt.plot([points[3], points[3]], [f1(points[3]), f2(points[3]-2*a)+1], 'k--', lw=.8)

plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$U(x)$', fontsize=18)
plt.xticks([])
plt.yticks([])
plt.legend(loc='best', frameon=0, fontsize=18)

plt.savefig('landauer.png', bbox_inches='tight')
plt.show()
plt.close()
```
